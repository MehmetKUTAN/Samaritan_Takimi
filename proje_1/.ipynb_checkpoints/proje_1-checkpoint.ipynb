```python
import pandas as pd
import xlsxwriter
import os
import time
from openpyxl import load_workbook
dosya_yolu=''   #colab'ta eğer çalıştıracak iseniz dosya yoluna '  sample_data/  '  ekleyiniz
```

```python
#############################################################################
def ortalama_hesapla(ogrenci_vizePuani,ogrenci_FinalPuani):
    ortalama=(int(ogrenci_vizePuani)*0.4)+(int(ogrenci_FinalPuani)*0.6)
    return ortalama
```

```python
#############################################################################
def ders_durumu(ortalama):
        print("*******Harf aralığı Harran üniversitesi'ne uygun olarak oluşturulmuştur******")
        state="Geçti"
        harf_notu="AA"
        if ortalama>=0 and ortalama<35:
                state="Kaldı"
                harf_notu="FF"
        elif ortalama>=35 and ortalama<45:
                state="Şartlı Geçti"
                harf_notu="DD"
        elif ortalama>=45 and ortalama<55:
                state="Şartlı Geçti"
                harf_notu="DC"
        elif ortalama>=55 and ortalama<65:
                state="Geçti"
                harf_notu="CC"
        elif ortalama>=65 and ortalama<75:
                state="Geçti"
                harf_notu="CB"
        elif ortalama>=75 and ortalama<85:
                state="Geçti"
                harf_notu="BB"
        elif ortalama>=85 and ortalama<90:
                state="Geçti"
                harf_notu="BA"
        elif ortalama>=90 and ortalama<=100:
                state="Geçti"
                harf_notu="AA"
        else:
                print("!! ortalama veya Sinav puanlarini kontrol ediniz")
                
        print("Ders ortalamanız: ",ortalama," Harf notunuz:",harf_notu," Ders Durumu:",state)
        print("*****************************************************************************")
        time.sleep(3)
        return state,harf_notu
```

```python
#############################################################################
def ogrenci_kontrol(okul_numara):
        if not os.path.exists(dosya_yolu+ders_adi+'_Dataframe.xlsx'):
                return False
        else:
                data = pd.read_excel(dosya_yolu+ders_adi+'_Dataframe.xlsx')
                if okul_numara in data.ogrenci_okulNo:
                        print("Ogrenci kaydı bulunmaktadır")
                        time.sleep(3)
                        return True
                return False
```

```python
#############################################################################
def ogrenci_bilgileri_function(ogrenci_sayisi,ders_adi):
    ogrenci_bilgileri=[]
    for i in range(ogrenci_sayisi):
        os.system('cls' if os.name == 'nt' else 'clear')
        check=False
        print(i+1,".Ogrenci Bilgileri")
        ogrenci_okulNo=int(input("Ogrenci okul no giriniz: "))
        check=ogrenci_kontrol(ogrenci_okulNo)
        if check is True:
                continue
        else:
                ogrenci_adi=input("Ogrenci adini giriniz: ")
                ogrenci_soyadi=input("ogrenci Soyadini giriniz: ")
                ogrenci_vizePuani=int(input("Ogrenci Vize notunu giriniz: "))
                ogrenci_FinalPuani=int(input("Ogrenci Final notunu giriniz: "))
                ortalama=ortalama_hesapla(ogrenci_vizePuani,ogrenci_FinalPuani)
                state,harf_notu=ders_durumu(ortalama)
                ogrenci_bilgileri.append(
                        {
                                "Ders":ders_adi,
                                "ogrenci_adi":ogrenci_adi,
                                "ogrenci_soyadi":ogrenci_soyadi,
                                "ogrenci_okulNo":ogrenci_okulNo,
                                "ogrenci_vizePuani":ogrenci_vizePuani,
                                "ogrenci_FinalPuani":ogrenci_FinalPuani,
                                "ortalama":ortalama,
                                "state":state,
                                "harfNotu":harf_notu,
                        }
                )
    return ogrenci_bilgileri
```

```python

#############################################################################
def write_excel(ogrenci_bilgileri,ders_adi):
        try:
                if os.path.exists(dosya_yolu+ders_adi+'_Dataframe.xlsx'):
                        data = pd.read_excel(dosya_yolu+ders_adi+'_Dataframe.xlsx')
                        size=len(data.ogrenci_okulNo)+1
                        workbook = load_workbook(filename=dosya_yolu+ders_adi+'_Dataframe.xlsx')
                        sheet = workbook.active
                        print(workbook)
                        newRowLocation=size+1
                        for row_num, data in enumerate(ogrenci_bilgileri):
                                sheet.cell(column=1,row=newRowLocation, value=ogrenci_bilgileri[row_num]["Ders"])
                                sheet.cell(column=2,row=newRowLocation, value=ogrenci_bilgileri[row_num]["ogrenci_adi"])
                                sheet.cell(column=3,row=newRowLocation, value=ogrenci_bilgileri[row_num]["ogrenci_soyadi"])
                                sheet.cell(column=4,row=newRowLocation, value=int(ogrenci_bilgileri[row_num]["ogrenci_okulNo"]))
                                sheet.cell(column=5,row=newRowLocation, value=ogrenci_bilgileri[row_num]["ogrenci_vizePuani"])
                                sheet.cell(column=6,row=newRowLocation, value=ogrenci_bilgileri[row_num]["ogrenci_FinalPuani"])
                                sheet.cell(column=7,row=newRowLocation, value=ogrenci_bilgileri[row_num]["ortalama"])
                                sheet.cell(column=8,row=newRowLocation, value=ogrenci_bilgileri[row_num]["state"])
                                sheet.cell(column=9,row=newRowLocation, value=ogrenci_bilgileri[row_num]["harfNotu"])
                        workbook.save(filename=dosya_yolu+ders_adi+'_Dataframe.xlsx')
                        workbook.close()
                        if os.path.exists(dosya_yolu+ders_adi+'_Dataframe.xlsx'):
                                print("Dataframe is Succesfull")
                        else:
                                print("Dataframe is UnSuccesfull")
                else:
                        with xlsxwriter.Workbook(dosya_yolu+ders_adi+'_Dataframe.xlsx') as workbook:
                                worksheet = workbook.add_worksheet()
                                for row_num, data in enumerate(ogrenci_bilgileri):
                                        worksheet.write_row(row_num, 0, data)
                                for row_num, data in enumerate(ogrenci_bilgileri):
                                        worksheet.write_row(row_num+1, 0, (ogrenci_bilgileri[row_num]["Ders"],
                                                                        ogrenci_bilgileri[row_num]["ogrenci_adi"],
                                                                        ogrenci_bilgileri[row_num]["ogrenci_soyadi"],
                                                                        ogrenci_bilgileri[row_num]["ogrenci_okulNo"],
                                                                        ogrenci_bilgileri[row_num]["ogrenci_vizePuani"],
                                                                        ogrenci_bilgileri[row_num]["ogrenci_FinalPuani"],
                                                                        ogrenci_bilgileri[row_num]["ortalama"], 
                                                                        ogrenci_bilgileri[row_num]["state"],
                                                                        ogrenci_bilgileri[row_num]["harfNotu"]))
                        if os.path.exists(dosya_yolu+ders_adi+'_Dataframe.xlsx'):
                                print("Dataframe is Succesfull")
                        else:
                                print("Dataframe is UnSuccesfull")
        except Exception as e:
                print("Hatali islemi. Program sonu")
                exit()         
```

```python
  
#############################################################################
if __name__ == '__main__':
        ders_adi=input("Ders adini giriniz: ")
        ogrenci_sayisi=int(input("ogrenci_sayisini giriniz: "))
        ogrenci_bilgileri=ogrenci_bilgileri_function(ogrenci_sayisi, ders_adi)
        write_excel(ogrenci_bilgileri,ders_adi)
```
